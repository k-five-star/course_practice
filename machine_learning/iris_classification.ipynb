iris datasets

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
```

```python
print(iris.feature_names) # prints the names of the four features
```

```python
print(iris.target) # 0, 1, 2 are the three classes of iris flowers (setosa, versicolor, virginica)
```

```python
import matplotlib.pyplot as plt

# Plotting the graph
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)

# Adding labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')

# Displaying the graph
plt.show() 
```

kNN training

```python
from sklearn.model_selection import train_test_split

X = iris.data # X is the features in our dataset
y = iris.target  # y is the Labels in our dataset

# Splitting the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4) # 80% training and 20% test

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=6) # K=6
knn.fit(X_train, y_train) # Training

y_pred = knn.predict(X_test) # Testing
scores = metrics.accuracy_score(y_test, y_pred) # Accuracy

print(scores)
```

kNN prediction

```python
knn = KNeighborsClassifier(n_neighbors=5) # K=5
knn.fit(X_train, y_train) # Training

classes = {0:'setosa', 1:'versicolor', 2:'virginica'} # Dictionary to map classes to names
# Making prediction on some unseen data
x_new = [[3, 4, 5, 2], [5, 4, 2, 2]]

y_predict = knn.predict(x_new)

print("for x_new_1 : ", classes[y_predict[0]])
print("for x_new_2 : ", classes[y_predict[1]])
```
