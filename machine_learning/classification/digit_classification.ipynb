## 1. ready

### a. show image

```python
import matplotlib.pyplot as plt # for plotting
from sklearn import datasets, metrics # for loading dataset and measuring accuracy
from sklearn.model_selection import train_test_split # for splitting dataset

digits = datasets.load_digits() # load dataset

plt.imshow(digits.images[0], cmap = plt.cm.gray_r, interpolation='nearest') # show the first image
```

### b. reshape images

```python
n_samples = len(digits.images) # number of images
data = digits.images.reshape((n_samples, -1)) # flatten the images -1 means the dimension is inferred
```

## 2. test

### a. separate train data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.5, shuffle=False) # split the dataset into training and testing sets
    # digits.target means the labels, shuffle means the data is not shuffled, test_size means the ratio of testing data

from sklearn.neighbors import KNeighborsClassifier # import KNN classifier

knn = KNeighborsClassifier() # create a KNN classifier
knn.fit(X_train, y_train) # train the classifier

y_pred = knn.predict(X_test) # predict the labels of testing data. y_pred is the predicted labels
scores = metrics.accuracy_score(y_test, y_pred) # calculate the accuracy. y_test is the true labels
print(scores)
```

### b. test

```python
plt.imshow(X_test[10].reshape(8,8), cmap = plt.cm.gray_r, interpolation='nearest') # show the 10th image as 8x8 matrix
y_pred = knn.predict([X_test[10]]) # predict the label of the 10th image the input must be a 2D array
print(y_pred) # print the predicted label
```
