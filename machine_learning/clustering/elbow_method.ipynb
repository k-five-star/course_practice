### 1. show datasets

```python
import matplotlib.pyplot as plt # Import matplotlib for plotting
import numpy as np # Import numpy for array manipulation
from sklearn.cluster import KMeans # Import KMeans from sklearn

X = np.array([
    [6, 3], [11, 15], [17, 12], [24, 10], [20, 25], [22, 30], 
    [85, 70], [71, 81], [60, 79], [56, 52], [81, 91], [80, 81]
]) # Data points
 
plt.scatter(X[:,0], X[:,1]) # Plot the data points
```

### 2. elbow method

```python
kmeans = [KMeans(n_clusters=i) for i in range(1, 10)] # Create a list of KMeans objects for each value of k
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))] # Calculate the inertia for each model 
    #inerita means the sum of squared distances of samples to their closest cluster center

plt.plot(range(1, 10), score) # Plot the elbow curve
plt.xlabel('Number of Clusters') # Label the x-axis
plt.ylabel('Score') # Label the y-axis 
plt.title('Elbow Curve') # Add a title
plt.show() # Show the plot

```
