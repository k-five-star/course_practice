### 1. show dataset

```python
import matplotlib.pyplot as plt # for plotting
import numpy as np # for array operations
from sklearn.cluster import KMeans # for K-means algorithm

X = np.array([[6, 3], [11, 15], [17, 12], [24, 10], [20, 25], 
              [22, 30], [85, 70], [71, 81], [60, 79], [56, 52], [81, 91], [80, 81]]) # data points

plt.scatter(X[:,0],X[:,1]) # plot data points
```

### 2. train and test

```python
kmeans = KMeans(n_clusters=2) # K-means with 2 clusters
kmeans.fit(X) # fit the data points

print(kmeans.cluster_centers_) # print the cluster centers

print(kmeans.labels_) # print the cluster labels

plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow') # plot data points with color
```
