### 1. show dataset

```python
import matplotlib.pyplot as plt # for plotting
import seaborn as sns; sns.set()  # for plot styling
import numpy as np # for array
from sklearn.cluster import KMeans # for KMeans clustering
from sklearn.datasets._samples_generator import make_blobs # for generating isotropic Gaussian blobs for clustering

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0) # generate isotropic Gaussian blobs for clustering
plt.scatter(X[:, 0], X[:, 1], s=50); # plot the data points
plt.show()
```

### 2. K-Means

```python
kmeans = KMeans(n_clusters=4) # create a KMeans object with 4 clusters
kmeans.fit(X) # fit the KMeans object to the data
y_kmeans = kmeans.predict(X) # predict the clusters for the data points

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis') # plot the data points with the predicted clusters

centers = kmeans.cluster_centers_ # get the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5); # plot the cluster centers

plt.show()
```
